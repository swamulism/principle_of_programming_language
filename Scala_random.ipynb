```scala
def facL(x: Int): Int = {
    if (x <= 0) {
        return 1
    } else {
        var prod = 1
        for (i <- 1 to x) {
            prod *= i
        }
        return prod
    }
}
```

```scala
facL(5)
```

```scala
def facR(x: Int): Int = {
    if (x <= 0) {
        1;
    } else {
        x * facR(x-1);
    }
}
```

```scala
facR(8)
```
