```scala
def facL(x: Int): Int = {
    if (x <= 0) {
        return 1
    } else {
        var prod = 1
        for (i <- 1 to x) {
            prod *= i
        }
        return prod
    }
}
```

```scala
def fib(n: Int): Int = {
    if (n <= 2) {
        return 1
    } else{
        return fib(n-1) + fib(n-2)
    }
}
```

```scala
def facNonTail(n: Int): Int = {
    if (n <= 1) {
        return 1
    } else {
        return n * facNonTail(n-1)
    }
}
```

```scala
def factTailRec(acc: Int, n: Int): Int = {
    if (n <= 1) {
        return acc
    } else {
        return factTailRec(acc*n, n-1)
    }
}

def factTail(n: Int): Int = {
    return factTailRec(1, n)
}
```

```scala
def sqrtTR(j: Int, n: Int): Int = {
    if (n <= 0) {
        return 0
    } else if (j * j < n) {
        return sqrtTR(j+1, n)
    } else {
        return j
    }
}

def sqrtRec(n: Int): Int = {
    return sqrtTR(1, n)
}
```

```scala
sqrtRec(4)
```

```scala
def fibCrazyRec(n: Int, j: Int = 1, acc1: Int = 0, acc2: Int = 1): Int = {
    if (n <= 2) {
        return 1
    } else if (j < n) {
        return fibCrazyRec(n, j+1, acc2, acc1 + acc2)
    } else {
        return acc2
    }
}
```

```scala
fibCrazyRec(10)
```

```scala
sealed trait List {
    def length(): Int
    def isEmpty(): Boolean
}
case object Nil extends List {
    def length(): Int = {
        0
    }
    def isEmpty(): Boolean = {true}
}
case class Cons(i: Int, l: List) extends List {
    def length(): Int = {
        1 + l.length()
    }
    def isEmpty(): Boolean = {false}
}
```

\[ 1, 2, 3, 5, 7 \]

```scala
val tmp = Cons(1, Cons(2, Cons(3, Cons(5, Cons(7, Nil)))))
```

```scala
tmp.length()
```

```scala
def lengthOfList(l: List, acc: Int = 0): Int = {
    l match {
        case Nil => acc
        case Cons(_, l1) => {
            lengthOfList(l1, acc + 1)
        } 
    }
}
```

```scala
lengthOfList(tmp)
```

```scala
def isAscendingOrder(lst: List): Boolean = lst match {
    case Nil => true
    case Cons(_, Nil) => true
    case Cons(i1, Cons(i2, l)) if (i1 > i2) => false 
    case Cons(il, lst1) => isAscendingOrder(lst1)
}
```
